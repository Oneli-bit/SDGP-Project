```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
female_data = pd.read_csv("finalizeDataSet.csv")
female_data
```

```python
female_data.head()
```

```python
print(female_data.columns)
```

```python
x = female_data.drop([" Fat","Obs"," BMI"],axis = 1)
y = female_data[' Fat']

from sklearn.feature_selection import mutual_info_regression
from sklearn.feature_selection import SelectKBest
```

```python
fs = SelectKBest(score_func=mutual_info_regression, k=12)
fs.fit(x,y)
```

```python
fs.scores_
```

```python
mi_score = pd.Series(fs.scores_, index = x.columns)
mi_score
```

```python
mi_score.sort_values(ascending=False).plot.bar(figsize=(6,4))
```

```python
x_selected = fs.fit_transform(x,y)
x_selected = pd.DataFrame(x_selected)
```
